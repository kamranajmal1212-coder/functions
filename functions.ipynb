```python
1 What is the difference between a function and a method in Python?
->In Python, a function is an independent block of code defined using def or lambda that performs
  a specific task and can be called directly by its name. A method, on the other hand,
  is a function that is defined inside a class and is associated with an object. Because of this,
  a method is always called using the object it belongs to, such as object.method().
  While functions stand on their own, methods operate on the data stored within an object and often use that object’s attribute

2 Explain the concept of function arguments and parameters in Python.
->In Python, parameters are the variable names listed in a function’s definition,
 while arguments are the actual values passed to the function when it is called.
 Parameters act as placeholders that receive the values supplied during the call.
 For example, in def add(a, b):, the variables a and b are parameters.
 When the function is executed using add(5, 3), the numbers 5 and 3 are
 the arguments assigned to those parameters. This system allows functions
 to accept input values, process them, and produce outputs, making them flexible and reusable.

3. What are the different ways to define and call a function in Python?
->In Python, a function can be defined in two main ways: using the traditional def keyword,
 which allows you to create a named function with a block of code, and using a lambda function,
 which creates a small anonymous function in a single line. Once defined, a function is called
 by writing its name followed by parentheses, optionally passing arguments inside them (for example, add(5, 3)).
 Methods inside classes are also functions but are called through objects, such as obj.method().
 Functions can also be called using keyword arguments, positional arguments, default arguments,
 or unpacking using *args and **kwargs, giving Python a flexible and powerful way to define and invoke functions.

4. What is the purpose of the `return` statement in a Python function?
->The return statement in a Python function is used to send a value back to the
part of the program where the function was called. It ends the function’s execution
and provides the final output of that function. Without a return statement, a function
automatically returns None, meaning it doesn’t give back any usable value. By using return,
you can produce results, pass data to other parts of the program, and make your functions
reusable and meaningful.

5. What are iterators in Python and how do they differ from iterables?
->In Python, an iterable is any object that can be looped over, such as lists,
tuples, strings, sets, or dictionaries. An iterable contains multiple elements
and provides an iterator when passed to the iter() function. An iterator, on the
other hand, is an object that allows you to retrieve elements one at a time using
the next() function. It keeps track of the current position in the sequence so it
knows which item to return next.

6. Explain the concept of generators in Python and how they are defined.
->In Python, generators are special functions that allow you to produce a sequence
of values one at a time, instead of generating and storing all values in memory at
once. They use the yield keyword instead of return. Each time a generator yields a
value, it pauses its execution and remembers its internal state. When the generator
is called again, it resumes from where it left off rather than starting from the beginning.
Generators are defined just like normal functions using the def keyword, but they contain
one or more yield statements.

7. What are the advantages of using generators over regular functions?
->Generators offer several advantages over regular functions, especially when
dealing with large amounts of data. The biggest benefit is memory efficiency—generators
produce values one at a time instead of creating and storing an entire list in memory,
which makes them ideal for large sequences or streams of data. They also support lazy
evaluation, meaning values are generated only when needed, which improves performance
and reduces unnecessary computations. Additionally, generators allow functions to pause
and resume their execution using yield, making them useful for tasks like reading large
files, handling infinite sequences, or managing pipelines of data. Overall, generators
help create cleaner, faster, and more efficient programs compared to regular functions
that return full lists or data structures at once.

8. What is a lambda function in Python and when is it typically used?
->A lambda function in Python is a small, anonymous function defined using the lambda
keyword instead of def. It can take any number of arguments but can contain only a
single expression, whose result is automatically returned. Because lambda functions
are concise and do not require a name, they are typically used for short, simple
operations that are needed only once—such as in map(), filter(), sorted(), or as
quick callback functions. For example, lambda x: x * 2 is a lambda function that
doubles its input. Lambdas help keep code shorter and cleaner when writing small,
temporary functions.

9. Explain the purpose and usage of the `map()` function in Python
->The map() function in Python is used to apply a specific function to each item
in an iterable (such as a list, tuple, or set) and return a new map object containing
the transformed results. It allows you to process or modify multiple elements without
writing an explicit loop. The general form is map(function, iterable). For example,
map(lambda x: x * 2, [1, 2, 3]) applies the lambda function to every element and
produces values 2, 4, 6. You can convert the result to a list using list(map_object)
if needed. Overall, map() makes your code cleaner and more efficient by applying a
function to an entire collection in a single step.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python.
->In Python, map(), reduce(), and filter() are functional programming tools, but each serves
a different purpose. map() applies a function to every element in an iterable and returns a
new iterable of transformed values—for example, doubling all numbers in a list. filter()
selects elements from an iterable based on a condition and returns only those that satisfy
the given test, such as keeping only even numbers. reduce(), which is available in the
functools module, repeatedly applies a function to pairs of elements in an iterable to
reduce the entire sequence into a single cumulative value—for example, summing all numbers
in a list. In short, map() transforms, filter() selects, and reduce() combines elements
into one final result.


11. Using pen & Paper write the internal mechanism for sum operation using  reduce function on this given
list:[47,11,42,13];
->Step 1:

Take the first two elements:
a = 47, b = 11
→ 47 + 11 = 58

Step 2:

Now the result becomes the new a and the next list element becomes b:
a = 58, b = 42
→ 58 + 42 = 100

Step 3:

Again, take the result as a and the next element as b:
a = 100, b = 13
→ 100 + 13 = 113
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total
# 2 Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(text):
    return text[::-1]
# 3  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_list(numbers):
    squared = []
    for num in numbers:
        squared.append(num * num)
    return squared
# 4  Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    """Check if a number n is prime."""
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# Example usage: check primes from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
# 5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total number of terms to generate
        self.count = 0          # Counter for terms generated
        self.a = 0              # First Fibonacci number
        self.b = 1              # Second Fibonacci number

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value
            self.count += 1
            return next_value

# Example usage:
n = 10
fib_iter = FibonacciIterator(n)
for num in fib_iter:
    print(num)
# 6  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    """Generate powers of 2 from 0 to max_exponent."""
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage:
n = 10  # maximum exponent
for value in powers_of_two(n):
    print(value)
# 7 Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):
    """Generator function to read a file line by line."""
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Remove trailing newline character
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")

# Example usage:
file_path = 'example.txt'  # Replace with your file path
for line in read_file_lines(file_path):
    print(line)
# 8  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Sample list of tuples
data = [(1, 5), (2, 2), (3, 8), (4, 1)]

# Using sorted() with a lambda function to sort by second element
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
# 9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Function to convert Celsius to Fahrenheit
def c_to_f(c):
    return (c * 9/5) + 32

# Using map() to convert the list
fahrenheit_temps = list(map(c_to_f, celsius_temps))

print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
# 10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Input string
input_string = "Hello, World!"

# Function to check if a character is not a vowel
def not_vowel(char):
    return char.lower() not in 'aeiou'

# Using filter() to remove vowels
filtered_chars = filter(not_vowel, input_string)

# Joining the characters back into a string
result_string = ''.join(filtered_chars)

print("Original string:", input_string)
print("Without vowels:", result_string)
# 11  Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

books = [
    ["Book A", 3, 250],  # [Title, Quantity, Price per unit]
    ["Book B", 2, 150],
    ["Book C", 5, 200]
]


#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
#order is smaller than 100,00 €.

# Sample list of orders: [order_number, quantity, price_per_item]
orders = [
    [1, 3, 20.0],   # 3 items × 20 € = 60 €
    [2, 5, 25.0],   # 5 items × 25 € = 125 €
    [3, 2, 40.0]    # 2 items × 40 € = 80 €
]

# Function to calculate total with 10€ surcharge if total < 100€
def process_orders(order_list):
    result = []
    for order in order_list:
        order_number, quantity, price = order
        total = quantity * price
        if total < 100:
            total += 10  # Add 10 € surcharge
        result.append((order_number, total))
    return result

# Call the function
processed_orders = process_orders(orders)
print(processed_orders)




#Write a Python program using lambda and map

# Sample list of orders: [order_number, quantity, price_per_item]
orders = [
    [1, 3, 20.0],   # 3 items × 20 € = 60 €
    [2, 5, 25.0],   # 5 items × 25 € = 125 €
    [3, 2, 40.0]    # 2 items × 40 € = 80 €
]

# Using map and lambda to calculate totals with 10€ surcharge if total < 100€
processed_orders = list(
    map(
        lambda order: (order[0], order[1]*order[2] + 10 if order[1]*order[2] < 100 else order[1]*order[2]),
        orders
    )
)

print(processed_orders)
```
